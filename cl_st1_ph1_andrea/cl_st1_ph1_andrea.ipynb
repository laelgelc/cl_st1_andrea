## Data Sampling

### Add transcript word counts

```python
from pathlib import Path
import re
import pandas as pd

transcripts_dir = Path("corpus/04_transcripts")

def count_transcript_words(commercial_id):
    transcript_path = transcripts_dir / f"{commercial_id}.txt"

    if not transcript_path.exists():
        return pd.NA

    transcript_text = transcript_path.read_text(encoding="utf-8")
    return len(re.findall(r"\b\w+\b", transcript_text))

successful_downloads_mask = tv_commercials_df["Download Success"].eq(True)

tv_commercials_df["Transcript Word Count"] = pd.NA

tv_commercials_df.loc[successful_downloads_mask, "Transcript Word Count"] = (
    tv_commercials_df.loc[successful_downloads_mask, "Commercial ID"]
    .apply(count_transcript_words)
)

tv_commercials_df["Transcript Word Count"] = (
    tv_commercials_df["Transcript Word Count"]
    .astype("Int64")
)

tv_commercials_df[["Commercial ID", "Download Success", "Transcript Word Count"]]
```

### Descriptive statistics and outliers for 1950 transcript word counts

```python
transcript_word_counts_1950_df = tv_commercials_df.loc[
    tv_commercials_df["Commercial ID"].str.contains(r"^tv_com_1950_", regex=True)
    & tv_commercials_df["Download Success"].eq(True),
    ["Commercial ID", "Title", "Transcript Word Count"]
].copy()

transcript_word_counts_1950 = transcript_word_counts_1950_df["Transcript Word Count"]

q1 = transcript_word_counts_1950.quantile(0.25)
q3 = transcript_word_counts_1950.quantile(0.75)
iqr = q3 - q1

lower_outlier_limit = q1 - 1.5 * iqr
upper_outlier_limit = q3 + 1.5 * iqr

transcript_word_count_outliers_1950 = transcript_word_counts_1950_df.loc[
    (transcript_word_counts_1950_df["Transcript Word Count"] < lower_outlier_limit)
    | (transcript_word_counts_1950_df["Transcript Word Count"] > upper_outlier_limit)
    ].sort_values("Transcript Word Count")

transcript_word_counts_1950.describe()
```

### Randomly select balanced non-outlier sample by decade

Given that the minimum number of non-outlier commercials, longer than 20 words, is 103, I'd like to randomly select 103 commercials from each decade. What is the best way to do this? I was thinking of creating a `Selected` column to indicate which commercials were selected...

Yes — creating a `Selected` column is a good approach because it keeps the sampling decision inside `tv_commercials_df`, makes the sample easy to inspect, and lets you save/export it with the rest of the metadata.

I would recommend:

1. Initialise `Selected` as `False`.
2. Filter to eligible commercials:
   - `Download Success == True`
   - `Outlier == False`
   - `Transcript Word Count > 20`
3. Infer decade from `Commercial ID`.
4. Randomly sample `103` rows per decade.
5. Mark those rows as `Selected == True`.
6. Use a fixed `random_state` so the selection is reproducible.

```python
sample_size_per_decade = 103
random_seed = 42

tv_commercials_df["Selected"] = False

eligible_commercials_df = (
    tv_commercials_df.loc[
        tv_commercials_df["Download Success"].eq(True)
        & tv_commercials_df["Outlier"].eq(False)
        & tv_commercials_df["Transcript Word Count"].gt(20)
        ]
    .assign(
        Decade_From_ID=lambda df: df["Commercial ID"].str.extract(r"^tv_com_(\d{4})_")[0]
    )
)

selected_commercial_indices = (
    eligible_commercials_df
    .groupby("Decade_From_ID", group_keys=False)
    .sample(n=sample_size_per_decade, random_state=random_seed)
    .index
)

tv_commercials_df.loc[selected_commercial_indices, "Selected"] = True

selected_commercials_by_decade = (
    tv_commercials_df.loc[tv_commercials_df["Selected"]]
    .assign(
        Decade_From_ID=lambda df: df["Commercial ID"].str.extract(r"^tv_com_(\d{4})_")[0]
    )
    .groupby("Decade_From_ID")
    .size()
    .reset_index(name="Selected Commercials")
)

selected_commercials_by_decade
```

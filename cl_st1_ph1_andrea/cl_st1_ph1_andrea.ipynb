## Import `TV_commercials_1950_2025.xlsx` into a pandas DataFrame

```python
import pandas as pd

tv_commercials_df = pd.read_excel(
    "corpus/00_sources/TV_commercials_1950_2025.xlsx",
    engine="openpyxl"
)
```

```python
tv_commercials_df = tv_commercials_df.drop(columns=["Unnamed: 7", "Unnamed: 8"])
```

## Add sequential video identifiers

```python
unique_video_ids = {
    url: f"video_{index:04d}"
    for index, url in enumerate(tv_commercials_df["URL"].drop_duplicates(), start=1)
}

video_id_position = tv_commercials_df.columns.get_loc("Category") + 1

tv_commercials_df.insert(
    video_id_position,
    "Video ID",
    tv_commercials_df["URL"].map(unique_video_ids)
)
```

## List repeated videos

```python
repeated_videos_df = (
    tv_commercials_df["Video ID"]
    .value_counts()
    .loc[lambda counts: counts > 1]
    .reset_index()
)

repeated_videos_df.columns = ["Video ID", "Occurrences"]

repeated_videos_df
```

## Add commercial identifiers

```python
commercial_id_position = tv_commercials_df.columns.get_loc("Category") + 1

tv_commercials_df.insert(
    commercial_id_position,
    "Commercial ID",
    (
            "tv_com_"
            + tv_commercials_df["Decade"].astype(str)
            + "_"
            + tv_commercials_df["Sequence"].astype(str)
    )
)

tv_commercials_df
```

## Save metadata as NDJSON

```python
tv_commercials_df.to_json(
    "corpus/00_sources/tv_commercials.ndjson",
    orient="records",
    lines=True,
    force_ascii=False
)
```

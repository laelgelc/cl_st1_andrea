# Corpus Linguistics - Study 1 - Phase 2 - Andrea

This document aims to organise the commercial verbal subcorpus for Lexical Multidimensional Analysis.

## Organise the commercial verbal subcorpus

### Import `tv_commercials.ndjson` into a pandas DataFrame

```python
import pandas as pd

tv_commercials_df = pd.read_json(
    "../cl_st1_ph1_andrea/corpus/00_sources/tv_commercials.ndjson",
    lines=True
)
```

### Organise selected commercial transcripts by decade

Copy the transcript files for the selected commercials into the commercial verbal subcorpus. Only rows marked as `Selected == True` are included. Each transcript is copied from the Phase 1 transcript directory to `corpus/commercial_verbal`, where files are organised into subdirectories named after their decade. The copied filenames are based on each commercial's `Commercial ID`.

```python
from pathlib import Path
import shutil

source_transcripts_dir = Path("../cl_st1_ph1_andrea/corpus/04_transcripts")
target_verbal_corpus_dir = Path("corpus/commercial_verbal")

selected_commercials_df = tv_commercials_df.loc[
    tv_commercials_df["Selected"].eq(True),
    ["Decade", "Commercial ID"]
].copy()

copied_transcripts = []
missing_transcripts = []

for _, commercial in selected_commercials_df.iterrows():
    decade = str(commercial["Decade"])
    commercial_id = commercial["Commercial ID"]

    source_transcript_path = source_transcripts_dir / f"{commercial_id}.txt"
    target_decade_dir = target_verbal_corpus_dir / decade
    target_transcript_path = target_decade_dir / f"{commercial_id}.txt"

    if source_transcript_path.exists():
        target_decade_dir.mkdir(parents=True, exist_ok=True)
        shutil.copy2(source_transcript_path, target_transcript_path)
        copied_transcripts.append(
            {
                "Decade": decade,
                "Commercial ID": commercial_id,
                "Source Path": source_transcript_path,
                "Target Path": target_transcript_path,
            }
        )
    else:
        missing_transcripts.append(
            {
                "Decade": decade,
                "Commercial ID": commercial_id,
                "Source Path": source_transcript_path,
            }
        )

copied_transcripts_df = pd.DataFrame(copied_transcripts)
missing_transcripts_df = pd.DataFrame(missing_transcripts)

copied_transcripts_by_decade = (
    copied_transcripts_df
    .groupby("Decade")
    .size()
    .reset_index(name="Copied Transcripts")
)

copied_transcripts_by_decade
```

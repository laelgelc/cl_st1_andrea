# Corpus Linguistics - Study 1 - Phase 5 - Andrea

This document aims to organise the commercial verbal, visual, and cross-modal data for Analysis of Variance (ANOVA) to detect diachronic variation in discourses.

## Constants

Define the constants.

```python
from pathlib import Path

PROJECT = "cl_st1_ph5_andrea"
VERBAL = "verbal"
VERBAL_ANOVA_FILE = Path(f"output_{PROJECT}_ANOVA/anova_{VERBAL}_overall.tsv")
VISUAL = "visual"
VISUAL_ANOVA_FILE = Path(f"output_{PROJECT}_ANOVA/anova_{VISUAL}_overall.tsv")
CROSS_MODAL = "crossmodal"
CROSS_MODAL_ANOVA_FILE = Path(f"output_{PROJECT}_ANOVA/anova_{CROSS_MODAL}_overall.tsv")
CROSS_MODAL_COMPOSITE = "crossmodal_composite"
CROSS_MODAL_COMPOSITE_ANOVA_FILE = Path(f"output_{PROJECT}_ANOVA/anova_{CROSS_MODAL_COMPOSITE}_overall.tsv")
OUTPUT_DIR = "../cl_st1_ph1_andrea/corpus/00_sources/tables/anova"
```

## Commercial Verbal Discourse ANOVA

### Load ANOVA results

```python
import pandas as pd

verbal_anova_results_df = pd.read_csv(VERBAL_ANOVA_FILE, sep="\t")

verbal_anova_results_df
```

### Prepare verbal ANOVA summary table

This cell converts the SAS overall ANOVA output for the verbal dimensions into a compact summary table. For each dependent verbal dimension, it keeps the model F statistic and p-value, computes \(R^2\) as the ratio between Model SS and Corrected Total SS, and adds the corresponding percentage of explained variance. Numeric values are formatted with four decimal places and leading zeros are omitted for values between -1 and 1.

```python
import pandas as pd


def format_decimal(value, digits=4):
    """Format numeric values with fixed decimals and omit leading zero."""
    formatted = f"{float(value):.{digits}f}"

    if formatted.startswith("0."):
        formatted = formatted[1:]
    elif formatted.startswith("-0."):
        formatted = "-" + formatted[2:]

    return formatted


def format_p_value(value, digits=4):
    """Format SAS p-values and omit leading zero where applicable."""
    value = str(value).strip()

    if value.startswith("<."):
        return value

    if value.startswith("<0."):
        return "<" + value[2:]

    try:
        return format_decimal(float(value), digits=digits)
    except ValueError:
        return value


verbal_model_df = (
    verbal_anova_results_df[
        verbal_anova_results_df["Source"].eq("Model")
    ]
    .copy()
)

verbal_corrected_total_df = (
    verbal_anova_results_df[
        verbal_anova_results_df["Source"].eq("Corrected Total")
    ][["Dependent", "SS"]]
    .rename(columns={"SS": "SSCorrected Total"})
    .copy()
)

verbal_anova_df = verbal_model_df.merge(
    verbal_corrected_total_df,
    on="Dependent",
    how="left",
)

verbal_anova_df["RSquare"] = (
        verbal_anova_df["SS"]
        / verbal_anova_df["SSCorrected Total"]
)

verbal_anova_df["RSquare %"] = (
        verbal_anova_df["RSquare"]
        * 100
)

verbal_anova_df = pd.DataFrame(
    {
        "Dim.": verbal_anova_df["Dependent"],
        "F": verbal_anova_df["FValue"].apply(format_decimal),
        "p": verbal_anova_df["ProbF"].apply(format_p_value),
        "RSquare": verbal_anova_df["RSquare"].apply(format_decimal),
        "RSquare %": verbal_anova_df["RSquare %"].apply(
            lambda value: format_decimal(value, digits=2)
        ),
    }
)

verbal_anova_df
```

### Export verbal ANOVA summary table to LaTeX

This cell exports the formatted verbal ANOVA summary table as a LaTeX table. The output is saved in the configured ANOVA tables directory with a descriptive caption and label for later inclusion in the manuscript or report.

```python
from pathlib import Path

output_dir = Path(OUTPUT_DIR)
output_dir.mkdir(parents=True, exist_ok=True)

verbal_anova_latex_path = output_dir / "verbal_anova.tex"

verbal_anova_latex_df = verbal_anova_df.rename(
    columns={
        "RSquare": r"R$^2$",
        "RSquare %": r"R$^2$ \%",
    }
)

verbal_anova_latex = verbal_anova_latex_df.to_latex(
    index=False,
    caption="Verbal Discourse ANOVA by Decade",
    label="tab:verbal_anova",
    column_format="lrrrr",
    escape=False,
)

verbal_anova_latex_path.write_text(verbal_anova_latex, encoding="utf-8")

verbal_anova_latex_path
```
